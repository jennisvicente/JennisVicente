# Top 3 countries: Main Sectors

```python
#Between $5-15million
df=df[(df['raised_amount_usd']>=5000000) & (df['raised_amount_usd']<=15000000)]

df.groupby(['country_code','main_sector'])['raised_amount_usd'].agg(['count', 'sum'])
```

```python
plt.figure(figsize= (16,14))
sns.set(style="whitegrid")

plt.subplot(2,1,1)
ax = sns.barplot(x="main_sector", y="raised_amount_usd", hue="country_code",estimator=np.sum, data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.title('Total Invested Amount (USD)')



plt.subplot(2,1,2)
ay= sns.countplot(x='main_sector' , hue='country_code', data=df)
ay.set_xticklabels(ay.get_xticklabels(), rotation=30)
plt.title('Number of Investments')
plt.show()
```
